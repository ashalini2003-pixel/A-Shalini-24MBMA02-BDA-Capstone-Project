```python
%sql
CREATE OR REPLACE TABLE `workspace`.`default`.`credit_card_fraud_cleaned` AS
SELECT
    -- Remove 'T' prefix and cast numeric part to INT safely
    TRY_CAST(REGEXP_REPLACE(Transaction_ID, '[^0-9]', '') AS INT) AS Transaction_ID,
    
    TRY_CAST(Time AS TIMESTAMP) AS Transaction_Time,
    TRY_CAST(Amount AS DOUBLE) AS Amount,
    
    TRIM(Card_Type) AS Card_Type,
    TRIM(Location) AS Location,
    TRY_CAST(Fraud_Flag AS INT) AS Fraud_Flag
FROM `workspace`.`default`.`credit_card_fraud_detection`;
```

```python
%sql
SELECT
    CASE 
        WHEN Amount > 5000 THEN 'High'
        WHEN Amount BETWEEN 1000 AND 5000 THEN 'Medium'
        ELSE 'Low'
    END AS Amount_Category,
    COUNT(*) AS total_transactions,
    SUM(CASE WHEN Fraud_Flag = 1 THEN 1 ELSE 0 END) AS fraud_transactions,
    ROUND(SUM(CASE WHEN Fraud_Flag = 1 THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) AS fraud_rate
FROM `workspace`.`default`.`credit_card_fraud_cleaned`
GROUP BY CASE 
            WHEN Amount > 5000 THEN 'High'
            WHEN Amount BETWEEN 1000 AND 5000 THEN 'Medium'
            ELSE 'Low'
         END
ORDER BY fraud_rate DESC;
```

```python
%sql
SELECT
    HOUR(Transaction_Time) AS transaction_hour,
    COUNT(*) AS total_transactions,
    SUM(CASE WHEN Fraud_Flag = 1 THEN 1 ELSE 0 END) AS fraud_transactions,
    ROUND(SUM(CASE WHEN Fraud_Flag = 1 THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) AS fraud_rate
FROM `workspace`.`default`.`credit_card_fraud_cleaned`
GROUP BY HOUR(Transaction_Time)
ORDER BY transaction_hour;
```

```python
%sql
SELECT
    DAYOFWEEK(Transaction_Time) AS day_of_week,
    COUNT(*) AS total_transactions,
    SUM(CASE WHEN Fraud_Flag = 1 THEN 1 ELSE 0 END) AS fraud_transactions,
    ROUND(SUM(CASE WHEN Fraud_Flag = 1 THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) AS fraud_rate
FROM `workspace`.`default`.`credit_card_fraud_cleaned`
GROUP BY DAYOFWEEK(Transaction_Time)
ORDER BY day_of_week;
```

```python
%sql
CREATE OR REPLACE TABLE `workspace`.`default`.`credit_card_fraud_features` AS
SELECT
    *,
    HOUR(Transaction_Time) AS Transaction_Hour,
    DAYOFWEEK(Transaction_Time) AS Transaction_Day,
    CASE 
        WHEN Amount > 5000 THEN 'High'
        WHEN Amount BETWEEN 1000 AND 5000 THEN 'Medium'
        ELSE 'Low'
    END AS Amount_Category
FROM `workspace`.`default`.`credit_card_fraud_cleaned`;
```

```python
%sql
SELECT 
    Amount_Category,
    Transaction_Hour,
    COUNT(*) AS total_txn,
    SUM(CASE WHEN Fraud_Flag = 1 THEN 1 ELSE 0 END) AS fraud_txn,
    ROUND(SUM(CASE WHEN Fraud_Flag = 1 THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) AS fraud_rate
FROM `workspace`.`default`.`credit_card_fraud_features`
GROUP BY Amount_Category, Transaction_Hour
ORDER BY fraud_rate DESC;
```

```python
%sql
SELECT
    COUNT(*) AS total_txn,
    SUM(CASE WHEN Fraud_Flag = 1 THEN 1 ELSE 0 END) AS fraud_txn,
    ROUND(SUM(CASE WHEN Fraud_Flag = 1 THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) AS fraud_rate,
    ROUND(AVG(Amount), 2) AS avg_amount,
    ROUND(AVG(CASE WHEN Fraud_Flag = 1 THEN Amount END), 2) AS avg_fraud_amount
FROM `workspace`.`default`.`credit_card_fraud_features`;
```

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set visualization style
sns.set(style="whitegrid", palette="muted", font_scale=1.1)

# Load dataset from SQL table into PySpark and then Pandas
df = spark.sql("SELECT * FROM `workspace`.`default`.`credit_card_fraud_detection`").toPandas()

# Show basic info
display(df.head())
df.info()
```

```python
# Print basic information
print("Total Transactions:", len(df))
print("Total Fraudulent Transactions:", df['Fraud_Flag'].sum())
print("Fraud Percentage:", round(df['Fraud_Flag'].mean() * 100, 2), "%")
```

```python
# Summary of numerical features
display(df.describe())

# Summary of categorical features
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    print(f"\nValue counts for {col}:")
    print(df[col].value_counts())
```

```python
# Check for missing values
print("Missing values per column:")
print(df.isnull().sum())

# Check data types
print("\nData types:")
print(df.dtypes)
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Fraud_Flag', data=df, palette='coolwarm')
plt.title("Distribution of Fraud vs Non-Fraud Transactions")
plt.xlabel("Fraud Flag (0 = Genuine, 1 = Fraudulent)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='Amount', hue='Fraud_Flag', bins=30, kde=True, palette='coolwarm')
plt.title("Transaction Amount Distribution by Fraud Status")
plt.xlabel("Transaction Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Card_Type', y='Fraud_Flag', data=df, estimator=lambda x: sum(x)/len(x)*100, palette='viridis')
plt.title("Fraud Rate by Card Type (%)")
plt.ylabel("Fraud Rate (%)")
plt.xlabel("Card Type")
plt.show()
```

```python
# Group by location and compute fraud rate
top_locations = (
    df.groupby('Location')['Fraud_Flag']
      .mean()
      .sort_values(ascending=False)
      .head(10)
      .reset_index()
)

plt.figure(figsize=(10,6))
sns.barplot(y='Location', x='Fraud_Flag', data=top_locations, palette='magma')
plt.title("Top 10 Locations by Fraud Rate (%)")
plt.xlabel("Fraud Rate (%)")
plt.ylabel("Location")
plt.show()
```

```python
# Convert Time column to datetime format if possible
df['Transaction_Hour'] = pd.to_datetime(df['Time'], errors='coerce').dt.hour

plt.figure(figsize=(10,6))
sns.lineplot(x='Transaction_Hour', y='Fraud_Flag', data=df, estimator='mean', marker='o', color='crimson')
plt.title("Fraud Rate by Hour of Transaction")
plt.xlabel("Hour of Day")
plt.ylabel("Average Fraud Rate")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(x='Fraud_Flag', y='Amount', data=df, palette='coolwarm')
plt.title("Transaction Amount vs Fraud Status")
plt.xlabel("Fraud Flag (0 = Genuine, 1 = Fraudulent)")
plt.ylabel("Transaction Amount")
plt.show()
```

```python
# Encode categorical variables temporarily for correlation calculation
numeric_df = df.copy()
for col in numeric_df.select_dtypes('object'):
    numeric_df[col] = pd.factorize(numeric_df[col])[0]

plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='Blues', fmt=".2f")
plt.title("Correlation Heatmap of Credit Card Features")
plt.show()
```
