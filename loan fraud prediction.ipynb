```python
%sql
-- ============================================================
SELECT 
  COUNT(*) AS total_applications,
  SUM(CASE WHEN `Default` = 1 THEN 1 ELSE 0 END) AS fraudulent_applications,
  ROUND(SUM(CASE WHEN `Default` = 1 THEN 1 ELSE 0 END) / COUNT(*) * 100, 2) AS fraud_percentage
FROM workspace.default.loan_fraud_prediction;
```

```python
%sql
-- STEP 6: FRAUD RATE BY EMPLOYMENT TYPE
-- ============================================================
SELECT 
  `Employment_Type`,
  COUNT(*) AS total_applications,
  SUM(CASE WHEN `Default` = 1 THEN 1 ELSE 0 END) AS fraud_cases,
  ROUND(SUM(CASE WHEN `Default` = 1 THEN 1 ELSE 0 END) / COUNT(*) * 100, 2) AS fraud_rate
FROM workspace.default.loan_fraud_prediction
GROUP BY `Employment_Type`
ORDER BY fraud_rate DESC;
```

```python
%sql
-- STEP 7: FRAUD RATE BY INCOME LEVEL
-- ============================================================
SELECT 
  CASE 
    WHEN `Income` < 30000 THEN 'Low Income (<30k)'
    WHEN `Income` BETWEEN 30000 AND 60000 THEN 'Medium Income (30k-60k)'
    WHEN `Income` BETWEEN 60001 AND 90000 THEN 'High Income (60k-90k)'
    ELSE 'Very High Income (>90k)'
  END AS income_category,
  COUNT(*) AS total_applications,
  SUM(CASE WHEN `Default` = 1 THEN 1 ELSE 0 END) AS fraud_cases,
  ROUND(SUM(CASE WHEN `Default` = 1 THEN 1 ELSE 0 END) / COUNT(*) * 100, 2) AS fraud_rate
FROM workspace.default.loan_fraud_prediction
GROUP BY 
  CASE 
    WHEN `Income` < 30000 THEN 'Low Income (<30k)'
    WHEN `Income` BETWEEN 30000 AND 60000 THEN 'Medium Income (30k-60k)'
    WHEN `Income` BETWEEN 60001 AND 90000 THEN 'High Income (60k-90k)'
    ELSE 'Very High Income (>90k)'
  END
ORDER BY fraud_rate DESC;
```

```python
%sql
-- STEP 8: FRAUD RATE BY CREDIT SCORE
-- ============================================================
SELECT 
  CASE 
    WHEN `Credit_Score` < 600 THEN 'Poor (<600)'
    WHEN `Credit_Score` BETWEEN 600 AND 700 THEN 'Fair (600-700)'
    WHEN `Credit_Score` BETWEEN 701 AND 750 THEN 'Good (700-750)'
    ELSE 'Excellent (>750)'
  END AS credit_category,
  COUNT(*) AS total_applications,
  SUM(CASE WHEN `Default` = 1 THEN 1 ELSE 0 END) AS fraud_cases,
  ROUND(SUM(CASE WHEN `Default` = 1 THEN 1 ELSE 0 END) / COUNT(*) * 100, 2) AS fraud_rate
FROM workspace.default.loan_fraud_prediction
GROUP BY 
  CASE 
    WHEN `Credit_Score` < 600 THEN 'Poor (<600)'
    WHEN `Credit_Score` BETWEEN 600 AND 700 THEN 'Fair (600-700)'
    WHEN `Credit_Score` BETWEEN 701 AND 750 THEN 'Good (700-750)'
    ELSE 'Excellent (>750)'
  END
ORDER BY fraud_rate DESC;
```

```python
%sql
-- STEP 9: FRAUD RATE BY LOAN PURPOSE
-- ============================================================
SELECT 
  `Loan_Purpose`,
  COUNT(*) AS total_applications,
  SUM(CASE WHEN `Default` = 1 THEN 1 ELSE 0 END) AS fraud_cases,
  ROUND(SUM(CASE WHEN `Default` = 1 THEN 1 ELSE 0 END) / COUNT(*) * 100, 2) AS fraud_rate
FROM workspace.default.loan_fraud_prediction
GROUP BY `Loan_Purpose`
ORDER BY fraud_rate DESC;
```

```python
%sql
-- STEP 10: FRAUD RATE BY PROPERTY AREA
-- ============================================================
SELECT 
  `Property_Area`,
  COUNT(*) AS total_applications,
  SUM(CASE WHEN `Default` = 1 THEN 1 ELSE 0 END) AS fraud_cases,
  ROUND(SUM(CASE WHEN `Default` = 1 THEN 1 ELSE 0 END) / COUNT(*) * 100, 2) AS fraud_rate
FROM workspace.default.loan_fraud_prediction
GROUP BY `Property_Area`
ORDER BY fraud_rate DESC;
```

```python
%sql
SELECT 
  `Education_Level`,
  COUNT(*) AS total_applications,
  SUM(CASE WHEN `Default` = 1 THEN 1 ELSE 0 END) AS fraud_cases,
  ROUND(
    SUM(CASE WHEN `Default` = 1 THEN 1 ELSE 0 END) / COUNT(*) * 100, 
    2
  ) AS fraud_rate
FROM workspace.default.loan_fraud_prediction
GROUP BY `Education_Level`
ORDER BY fraud_rate DESC;
```

```python
%sql
-- STEP 12: AVERAGE LOAN AMOUNT (Fraud vs Non-Fraud)
-- ============================================================
SELECT 
  CASE WHEN `Default` = 1 THEN 'Fraudulent' ELSE 'Legitimate' END AS loan_status,
  ROUND(AVG(`Loan_Amount `), 2) AS avg_loan_amount,
  ROUND(MAX(`Loan_Amount `), 2) AS max_loan_amount,
  ROUND(MIN(`Loan_Amount `), 2) AS min_loan_amount
FROM workspace.default.loan_fraud_prediction
GROUP BY `Default`;
```

```python
%sql
-- STEP 13: TOP 10 HIGH-VALUE FRAUDULENT LOANS
-- ============================================================
SELECT 
  `Customer_ID`,
  `Gender`,
  `Income`,
  `Loan_Amount `,
  `Credit_Score`,
  `Loan_Purpose`,
  `Employment_Type`,
  `Property_Area`
FROM workspace.default.loan_fraud_prediction
WHERE `Default` = 1
ORDER BY `Loan_Amount ` DESC
LIMIT 10;
```
