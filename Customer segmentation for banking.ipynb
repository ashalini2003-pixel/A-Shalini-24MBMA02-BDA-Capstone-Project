```python
%sql
CREATE OR REPLACE TABLE workspace.default.customer_segmentation_clean AS
SELECT
  TRIM(Customer_ID) AS Customer_ID,
  TRY_CAST(Age AS INT) AS Age,
  UPPER(TRIM(Gender)) AS Gender,
  TRY_CAST(REPLACE(`Income`, ',', '') AS DOUBLE) AS Income,
  TRY_CAST(REPLACE(`Account_Balance `, ',', '') AS DOUBLE) AS Account_Balance,
  TRY_CAST(Credit_Score AS INT) AS Credit_Score,
  TRY_CAST(Num_Products AS INT) AS Num_Products,
  TRY_CAST(Transaction_Frequency AS INT) AS Transaction_Frequency,
  TRY_CAST(REPLACE(`Avg_Transaction_Amount `, ',', '') AS DOUBLE) AS Avg_Transaction_Amount,
  TRY_CAST(REPLACE(`Loan_Amount`, ',', '') AS DOUBLE) AS Loan_Amount,
  TRY_CAST(Tenure_Years AS INT) AS Tenure_Years,
  UPPER(TRIM(City)) AS City,
  UPPER(TRIM(Channel_Preference)) AS Channel_Preference,
  UPPER(TRIM(Is_Active)) AS Is_Active
FROM workspace.default.customer_segmentation_for_banking
WHERE Customer_ID IS NOT NULL;
```

```python
%sql
SELECT
  COUNT(*) AS Total_Customers,
  ROUND(AVG(Age),2) AS Avg_Age,
  ROUND(AVG(Income),2) AS Avg_Income,
  ROUND(AVG(`Account_Balance `),2) AS Avg_Balance,
  ROUND(AVG(Credit_Score),2) AS Avg_Credit_Score
FROM workspace.default.customer_segmentation_for_banking;
```

```python
%sql
CREATE OR REPLACE TABLE workspace.default.customer_features AS
SELECT
  Customer_ID,
  Age,
  Gender,
  Income,
  Account_Balance,
  Credit_Score,
  Num_Products,
  Transaction_Frequency,
  Avg_Transaction_Amount,
  Loan_Amount,
  Tenure_Years,
  City,
  Channel_Preference,
  Is_Active,

  -- Group customers by income level
  CASE 
      WHEN Income < 300000 THEN 'Low Income'
      WHEN Income BETWEEN 300000 AND 700000 THEN 'Middle Income'
      ELSE 'High Income'
  END AS Income_Group,

  -- Group customers by age
  CASE 
      WHEN Age < 30 THEN 'Young'
      WHEN Age BETWEEN 30 AND 50 THEN 'Middle Aged'
      ELSE 'Senior'
  END AS Age_Group,

  -- Calculate total spend estimate
  ROUND(Transaction_Frequency * Avg_Transaction_Amount, 2) AS Total_Spend,

  -- Create a simple Financial Stability Index
  ROUND((Account_Balance + Income) / NULLIF(Loan_Amount, 0), 2) AS Financial_Stability_Index

FROM workspace.default.customer_segmentation_clean;
```

```python
%sql
SELECT
  Income_Group,
  Age_Group,
  ROUND(AVG(Total_Spend), 2) AS Avg_Spend,
  ROUND(AVG(Credit_Score), 0) AS Avg_Credit_Score,
  ROUND(AVG(Financial_Stability_Index), 2) AS Avg_Stability,
  COUNT(*) AS Total_Customers
FROM workspace.default.customer_features
GROUP BY Income_Group, Age_Group
ORDER BY Avg_Spend DESC;
```

```python
# Load the feature-engineered table into a Spark DataFrame
df_spark = spark.table("workspace.default.customer_features")

# Convert to Pandas (for visualization and quick checks)
df = df_spark.toPandas()
df.head()
```

```python
# Select only numeric columns for clustering
numeric_cols = ['Age', 'Income', 'Account_Balance', 'Credit_Score', 
                'Num_Products', 'Transaction_Frequency', 
                'Avg_Transaction_Amount', 'Loan_Amount', 
                'Total_Spend', 'Financial_Stability_Index']

# Drop rows with missing or infinite values
df = df.replace([np.inf, -np.inf], np.nan).dropna(subset=numeric_cols)
```

```python
# Income vs Account Balance
sns.scatterplot(x='Income', y='Account_Balance', hue='Age_Group', data=df)
plt.title('Income vs Account Balance by Age Group')
plt.show()

```

```python
# Average Credit Score by Income Group
sns.barplot(x='Income_Group', y='Credit_Score', data=df, estimator=np.mean, ci=None)
plt.title('Average Credit Score by Income Group')
plt.show()

```

```python
# Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Key Features')
plt.show()
```
